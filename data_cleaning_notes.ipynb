```python
import pandas as pd
import numpy as np

data_df = pd.read_csv('JEOPARDY_DATA.csv')
data_df.head(10)
```

```python
data_df.columns
# Minute promblems that will cause issues in later code
```

### Avoiding Spaces

```python
data_df.columns = data_df.columns.str.replace(" ",'')
data_df.columns
```

### Data Types

```python
data_df.dtypes
```

```python
# m/d/y is the order the data currently is in before manipulation
data_df['AirDate'] = pd.to_datetime(data_df['AirDate'], format='%m/%d/%Y') #capital Y means 4 digits
# preferable to use yyyy/mm/dd, easier to sort and funnel
data_df.head()
data_df.dtypes
```

### Assign (Using Lambda)

```python
# used to 
data_df = data_df.assign(month = lambda x : x['AirDate'].dt.month)
data_df.head()
```

```python
data_df['month2'] = data_df.apply(lambda x : x['AirDate'].month, axis = 1)
data_df.head()
```

### Filters

```python
#We might 
value_list = data_df['Value'].tolist()
value_list = value_list[0:50]
value_list
```

```python
# working with data but have to perserve original values
filtered_list = list(filter(lambda num: int(num.replace("$","")
                                       .replace(" ","").replace(",",""))>300
                            ,value_list))
filtered_list
```
